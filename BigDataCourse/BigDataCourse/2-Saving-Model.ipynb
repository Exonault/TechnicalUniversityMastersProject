# Saving a basic model

import fastai

```python
from fastai.vision.all import *
```

download and decompress a set with pics of cats and dogs

```python
path = untar_data(URLs.PETS)/"images"
```

```python
path
```

Define a function to label the images. In this case cats has capital first letter

```python
def is_cat(x): return x[0].isupper()
```

Data loaders

```python
dls = ImageDataLoaders.from_name_func(".",
                                     get_image_files(path), valid_pct=0.2, seed=42,
                                     label_func = is_cat,
                                     item_tfms = Resize(192))
```

```python
dls.show_batch(max_n=6) 
```

In this case true means it is a cat, false means it is a dog

```python
learner = vision_learner(dls, resnet18, metrics=error_rate)
learner.fine_tune(3)
```

```python
learner.export('modelCatOrDog.pkl')
```
