## Зареждане на данни

```python
path = "files/"
```

```python
import numpy as np
import pandas as pd
from pandas import read_excel

daysDf = read_excel(path + "day_project.xls", sheet_name="drujba", header=1, dtype={
    "O3": np.float64,
    "PM10": np.float64
})

daysDf
```

```python
columns = ['Date', 'NO', 'NO2', 'AirTemp', 'Press', 'UMR', 'WD', 'WS']
hoursDf = read_excel(path + "hour_project.xls", sheet_name="drujba", header=1,
                     usecols=columns)

hoursDf
```

## Сливане на данните и data cleaning

```python
data_df = pd.merge(hoursDf, daysDf, on='Date')
data_df
```

```python
data_df = data_df.dropna()
data_df = data_df.round({'NO': 2, "NO2": 2, "O3": 2, "RM10": 2, 'AirTemp': 1, "UMR": 1, 'Press': 0})
data_df
```

Разделяне на данните за обучение и тест

```python
X = data_df[["AirTemp", "Press", "UMR"]]
Y = data_df[["NO", "NO2", "O3", "RM10"]]
```

За трансформация на данните ще използваме `RobustScaller`

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler().fit(X)
X_scaled = scaler.transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

За обучение ще използваме `RandomForestRegressor`. Той е базиран на RandomForest който работи като взима средната стойност на `n` на брой дървета за решение (Decion Trees)

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(random_state=42)
regressor.fit(X_train, Y_train)
```

```python
print(regressor.score(X_test, Y_test))
```

```python
import joblib

joblib.dump(regressor, 'project_model.joblib')
joblib.dump(scaler, 'project_scaler.joblib')
```

# Визуален интерфейс

```python
from tkinter import *
from tkinter import messagebox
from joblib import load
import numpy as np

trained_regressor = load('project_model.joblib')
trained_scaler = load('project_scaler.joblib')
```

```python
root = Tk()
root.title("Predict Value")
root.resizable(False, False)

tempLabel = Label(root, text="Temperature (C)")
humidityLabel = Label(root, text="Humidity (%)")
pressureLabel = Label(root, text="Pressure (hPa)")
noLabel = Label(root, text="NO (ug/m3)")
no2Label = Label(root, text="NO2 (ug/m3)")
ozoneLabel = Label(root, text="Ozone (ug/m3)")
pm10Label = Label(root, text="PM10 (ug/m3)")

nonormL = Label(root, text="NO/NO2 norm is: %d" % 200)
ozoneNormL = Label(root, text="Ozone norm is: %d" % 200)
pm10NormL = Label(root, text="PM10 norm is : %d" % 50)

tempEntry = Entry(root)
humidityEntry = Entry(root)
pressureEntry = Entry(root)
noEntry = Entry(root, state='disabled')
no2Entry = Entry(root, state='disabled')
ozoneEntry = Entry(root, state='disabled')
pm10Entry = Entry(root, state='disabled')

tempLabel.grid(row=0, column=0)
tempEntry.grid(row=0, column=1)

humidityLabel.grid(row=1, column=0)
humidityEntry.grid(row=1, column=1)

pressureLabel.grid(row=2, column=0)
pressureEntry.grid(row=2, column=1)

noLabel.grid(row=3, column=0)
noEntry.grid(row=3, column=1)
nonormL.grid(row=3, column=2)

no2Entry.grid(row=4, column=0)
no2Entry.grid(row=4, column=1)

ozoneLabel.grid(row=5, column=0)
ozoneEntry.grid(row=5, column=1)
ozoneNormL.grid(row=5, column=2)

pm10Label.grid(row=6, column=0)
pm10Entry.grid(row=6, column=1)
pm10NormL.grid(row=6, column=2)
```

```python
def predict():
    noEntry.configure(state='normal')
    noEntry.delete(0, END)
    noEntry.configure(state='disabled')

    no2Entry.configure(state='normal')
    no2Entry.delete(0, END)
    no2Entry.configure(state='disabled')

    ozoneEntry.configure(state='normal')
    ozoneEntry.delete(0, END)
    ozoneEntry.configure(state='disabled')

    pm10Entry.configure(state='normal')
    pm10Entry.delete(0, END)
    pm10Entry.configure(state='disabled')

    try:
        arr = np.array([[float(tempEntry.get()), float(humidityEntry.get()), float(pressureEntry.get())]])
        arrScaled = trained_scaler.transform(arr)
        result = trained_regressor.predict(arrScaled)

        noEntry.configure(state='normal')
        noEntry.insert(0, round(result[0][0], 2))
        noEntry.configure(state='disabled')

        no2Entry.configure(state='normal')
        no2Entry.insert(0, round(result[0][1], 2))
        no2Entry.configure(state='disabled')

        ozoneEntry.configure(state='normal')
        ozoneEntry.insert(0, round(result[0][2], 2))
        ozoneEntry.configure(state='disabled')

        pm10Entry.configure(state='normal')
        pm10Entry.insert(0, round(result[0][3], 2))
        pm10Entry.configure(state='disabled')

    except ValueError:
        messagebox.showerror("Wrong value", "Enter float values")
```

```python
b = Button(root, text="Predict", command=predict)
b.grid(row=7, column=1)
```

```python
root.mainloop()
```

# Експериментиране с хиперпараметрите
`RandomForestRegressor` използва няколко хиперпараметъра <br>
- `n_estimators` - брой на дърветата в гората
- `max_depth` - максимална дълбочина на отделните дървета
- `min_samples_split`- минималният брой проби, които са необходими, за да бъде "разделен" възел.
- `min_samples_leaf`- минималният брой проби, който трябва да има всеки листен възел.
- `criterion` - функцията за изчисляване качеството при разделяне; поддържат се 3 опции: `squared_error`, `absolute_error`, `poisson`

<br>
За намиране на идеални хиперпараметри ще използваме `GridSearchCV` и ще тестваме различни комбинации от хиперпараметри


```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor


param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['squared_error', 'absolute_error', 'poisson']
}

random_forest = RandomForestRegressor(random_state=42)


grid_search = GridSearchCV(
    estimator=random_forest,
    param_grid=param_grid,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, Y_train)

print("Най-добри параметри:", grid_search.best_params_)

best_model = grid_search.best_estimator_


predictions = best_model.predict(X_test)
mse = mean_squared_error(Y_test, predictions)
print("Mean Squared Error върху тестовите данни:", mse)
```
