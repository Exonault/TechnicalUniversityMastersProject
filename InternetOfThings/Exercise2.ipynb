```python
import os
import numpy as np
import pandas as pd
from pandas import read_excel
```

```python
# os.chdir(".")
path = "files/"
print(path)
```

```python
daysDf = read_excel(path + "day.xls", sheet_name='pavlovo', header=1, dtype={'03': np.float64, 'PM10': np.float64})
daysDf
```

```python
hoursDf = read_excel(path + "hour.xls", sheet_name='pavlovo', header=1, usecols=['Date', 'NO', 'NO2', 'AirTemp', 'Press', 'UMR', 'WD', 'WS'])
hoursDf
```

```python
data_df = pd.DataFrame()

for i in range(daysDf.index.size):
   hour_data = hoursDf[hoursDf['Date'].str.contains(daysDf['Date'][i])]
   hour_data.insert(6, 'O3', daysDf['O3'][i])
   hour_data.insert(7, 'PM10', daysDf['PM10'][i])
   data_df = data_df._append(hour_data)

data_df
```

```python
data_df = pd.merge(hoursDf, daysDf, on='Date')
data_df
```

```python
data_df = data_df.dropna()
data_df
```

```python
data_df = data_df.round({'NO': 2, "NO2": 2, "O3": 2, "PM10": 2, 'AirTemp': 1, "UMR": 1, 'Press': 0})
data_df
```

```python
columns = ["NO", "NO2", "AirTemp", "Press", "UMR", "WD", "WS", "O3", "PM10"]
data_df.to_csv("data_ex2.csv", columns=columns)
```
